```python
def get_eandt(array):
    event_letter = "E"
    time_letter = "T"
    re_l = r"..[A-Z]...$"
    re_n = r'.. '

    i = 0
    j = 0
    k = 0
    event_data = []
    time_data = []
    for row in array:
        if re.search(re_l, str(row)):
            marker = str(row) 
            j = i
            k = i
            if re.match(marker, event_letter):
                j = j + 1
                while re.search(re_n, str(array[j])):
                    event_data.append((str(array[j])))
                    j = j + 1
            if re.match(marker, time_letter):
                k = k + 1
                while re.search(re_n, str(array[k])):
                    time_data.append((str(array[k])))
                    k = k + 1
            else:
                i = i + 1
        else:
            i = i + 1
            
    re_n = r'^\d(\d|\.)(\d|\.)(\d|\.)'
    event_data_list = []
    for row in event_data:
        data_set_split = row.split(" ")
        for line in data_set_split:
            if re.search(re_n, line):
                event_data_list.append(float(line[0:(len(line)-2)]))
                
    re_n = r'^\d(\d|\.)(\d|\.)(\d|\.)'
    time_data_list = []
    for row in time_data:
        data_set_split = re.split(" |'", row)
        for line in data_set_split:
            if re.search(re_n, line):
                time_data_list.append(float(line[0:(len(line)-1)]))

    final_data = np.column_stack((event_data_list, time_data_list))
    
    return final_data
```

```python
def CS_and_ITI_time(data, CS_s = 22, CS_e = 23, ITI_s = 15, ITI_e = 16):
    time = {}
    ITI_start = []
    ITI_end = []
    for row in data:
        if row[0] == ITI_s:
            ITI_start.append(row[1])
        if row[0] == ITI_e:
            ITI_end.append(row[1]) 

    CS_start = []
    CS_end = []
    for row in data:
        if row[0] == CS_s:
            CS_start.append(row[1])
        if row[0] == CS_e:
            CS_end.append(row[1])

    #names = ['ITI_start', 'ITI_end', 'CS_start', 'CS_end']
    #data = [ITI_start, ITI_end, CS_start, CS_end]
    #i = 0
    #for name in names:
        #time[name] = data[i]
        #i = i + 1
    
    data = np.array([ITI_start, ITI_end, CS_start, CS_end])
    col_names = ['ITI_start', 'ITI_end', 'CS_start', 'CS_end']
    df_times = pd.DataFrame(data.reshape(1,4), index = {session_list[0]}, columns = col_names)
    
    return df_times
```

we routinely measure 
(a) the number of lever deflections and food cup head entries during the CS period, 
count number of lever pressess, count number of food cup head entries

(b) the probability of contacting the lever or
entering the food cup during the CS period on each trial (defined
as the number of trials with a lever press or food cup entry, divided
by the total number of trials). count numnber of trials a lp or he occured then divide by number of trials

(c) the ave latency to the first lever
deflection or food cup entry. find latency for first lp or he for each trial then average over number of trials

```python
def get_CS_npl(data, time, right_lp = 1, left_lp = 2, he = 6):
    
    CS_lp = 0
    CS_lp_prob = 0
    CS_lp_latency_ave = 0
    
    CS_lp_times = {}
    CS_lp_latency = {}
    
    lp_trial_counter = 0
    CS_lp_latency_sum = 0
    
    CS_he = 0
    CS_he_prob = 0
    CS_he_latency_ave = 0
    
    CS_he_times = {}
    CS_he_latency = {}
        
    he_trial_counter = 0
    CS_he_latency_sum = 0

    i = 0
    while i < len(time['CS_start']):
        j = 0
        for row in data:
            if row[1] >= time['CS_start'][i] and row[1] <= time['CS_end'][i]:
                if row[0] == right_lp:
                    CS_lp = CS_lp + 1
                    CS_lp_times[row[1]] = row[0]
                    if j < 1:
                        j = j + 1
                        lp_trial_counter = lp_trial_counter + 1
                        CS_lp_latency[row[1]] = row[1] - time['CS_start'][i]
                        CS_lp_latency_sum = CS_lp_latency_sum + CS_lp_latency[row[1]]
                if row[0] == left_lp:
                    CS_lp = CS_lp + 1
                    CS_lp_times[row[1]] = row[0]
                    if j < 1:
                        j = j + 1
                        lp_trial_counter = lp_trial_counter + 1
                        CS_lp_latency[row[1]] = row[1] - time['CS_start'][i]
                        CS_lp_latency_sum = CS_lp_latency_sum + CS_lp_latency[row[1]]
                if row[0] == he:
                    CS_he = CS_he + 1
                    CS_he_times[time['CS_start'][i]] = row[1]
                    if j < 1:
                        j = j + 1
                        he_trial_counter = he_trial_counter + 1
                        CS_he_latency[row[1]] = row[1] - time['CS_start'][i]
                        CS_he_latency_sum = CS_he_latency_sum + CS_he_latency[row[1]]
                
        i = i + 1
    
    CS_lp_prob = lp_trial_counter / len(time['CS_start'])
    if CS_lp_latency_sum == 0.0:
        CS_lp_latency_ave = 10
    else:
        CS_lp_latency_ave = CS_lp_latency_sum / lp_trial_counter
            
    CS_he_prob = he_trial_counter / len(time['CS_start'])
    if CS_he_latency_sum == 0.0:
        CS_he_latency_ave = 10
    else:
        CS_he_latency_ave = CS_he_latency_sum / he_trial_counter
    
    npl = {}
    names = ['CS_lp', 'CS_lp_times', 'CS_lp_prob', 'CS_lp_latency_ave', 'CS_lp_latency', 'CS_he', 'CS_he_times', 'CS_he_prob', 'CS_he_latency_ave', 'CS_he_latency']
    data = [CS_lp, CS_lp_times, CS_lp_prob, CS_lp_latency_ave, CS_lp_latency, CS_he, CS_he_times, CS_he_prob, CS_he_latency_ave, CS_he_latency]
    i = 0
    for name in names:
        npl[name] = data[i]
        i = i + 1
        
    #names = ['CS_lp', 'CS_lp_times', 'CS_lp_prob', 'CS_lp_latency_ave', 'CS_lp_latency', 'CS_he', 'CS_he_times', 'CS_he_prob', 'CS_he_latency_ave', 'CS_he_latency']
    #data = [CS_lp, CS_lp_times, CS_lp_prob, CS_lp_latency_ave, CS_lp_latency, CS_he, CS_he_times, CS_he_prob, CS_he_latency_ave, CS_he_latency]
    #npl_table = pd.DataFrame(data)
    #npl_table_t = npl_table.T
    #npl_table_t.columns = names
    
    return npl
```

Using these measures we devised a
‘‘PCA Score’’, as follows. From the CS lever deflections and CS
food cup entries, we calculated three measures of approach: 

1) Response Bias (ratio of lever presses and food cup entries in relation
to total number of responses), 

2) Probability Difference (the difference
between the probability of pressing the lever and the probability of
entering the food cup), and 

3) Latency Score (difference between the
latencies to approach the lever and the food cup). 

Averaging these
three measures of approach produces a PCA Score for each
individual, on each day of training. 

Scores derived this way range
from 21 to +1, whereby scores of 21 and +1 indicate strong
biases toward goal-tracking and sign-tracking, respectively, and a
score of zero indicates that the two responses are equally
distributed. Table 1 describes these calculations in detail. 

We also
measured the number of food cup head entries during the intertrial
interval (ITI), but these are not included as part of the PCA
Score.

```python
def get_PCA_score(data):
    if abs(data['CS_lp']) > 0 | abs(data['CS_he']) > 0:
        response_bias = (data['CS_lp'] - data['CS_he']) / (data['CS_lp'] + data['CS_he'])
    else:
        response_bias = 0
    prob_diff = data['CS_lp_prob'] - data['CS_he_prob']
    lat_score = (data['CS_he_latency_ave'] - data['CS_lp_latency_ave']) / 10
    PCA_score = (response_bias + prob_diff + lat_score) / 3
    
    data = np.array([response_bias, prob_diff, lat_score, PCA_score])
    col_names = ['response_bias', 'prob_diff', 'lat_score', 'PCA_score']
    df_scores = pd.DataFrame(data.reshape(1,4), columns = col_names)
        
    return df_scores
```
