Data transformation. Goal of script is to pull in Med-PC files and extract event times and stamps into a dictonary with the session name as the key [event stamp, event time] as value. Directory structure is parent directory of complete experiment (here a pavlovian conditioning experiment) containing directories by session name (pav1, pav2, ext1, etc), within each session directory is a file for each animal for that session (this is the med-pc file we want to perform data transformation on)

Import necessary modles

```python
import numpy as np
import pandas as pd
import re
import os
```

```python
def get_eandt(array):
    event_letter = "E"
    time_letter = "T"
    re_l = r"..[A-Z]...$"
    re_n = r'.. '

    i = 0
    j = 0
    k = 0
    event_data = []
    time_data = []
    for row in array:
        if re.search(re_l, str(row)):
            marker = str(row) 
            j = i
            k = i
            if re.match(marker, event_letter):
                j = j + 1
                while re.search(re_n, str(array[j])):
                    event_data.append((str(array[j])))
                    j = j + 1
            if re.match(marker, time_letter):
                k = k + 1
                while re.search(re_n, str(array[k])):
                    time_data.append((str(array[k])))
                    k = k + 1
            else:
                i = i + 1
        else:
            i = i + 1
            
    re_n = r'^\d(\d|\.)(\d|\.)(\d|\.)'
    event_data_list = []
    for row in event_data:
        data_set_split = row.split(" ")
        for line in data_set_split:
            if re.search(re_n, line):
                event_data_list.append(float(line[0:(len(line)-2)]))
                
    re_n = r'^\d(\d|\.)(\d|\.)(\d|\.)'
    time_data_list = []
    for row in time_data:
        data_set_split = re.split(" |'", row)
        for line in data_set_split:
            if re.search(re_n, line):
                time_data_list.append(float(line[0:(len(line)-1)]))

    final_data = np.column_stack((event_data_list, time_data_list))
    
    return final_data
```

Create list of names of sessions in directory of session files

```python
session_dir_names = os.listdir(root.directory)
print(session_dir_names)
```

Loop through session list and create path name for each session directory

```python
session_dir_paths = []
for session in session_dir_names:
    session_dir_paths.append(root.directory + "/" + session)

session_dir_paths
```

Use os module and loop through session_paths to create list of sessions (session_list) (acorss all sessions and all animals) and a list of session data paths (session_data_paths)

```python
session_file_names = []
data_dic = {}

for path in session_dir_paths:
    sessions = os.listdir(path)
    for session in sessions:
        session_file_names.append(session)
        session_data_path = str(path) + "/" + str(session)
        data = pd.read_table(session_data_path)
        data_table = pd.DataFrame(data = data)
        data_array = data_table.values
        data_dic[session] = data_array

session_file_names
```

```python
i = 0
e_and_t = {}
while i < len(session_file_names):
    eandt = get_eandt(data_dic[session_file_names[i]])
    e_and_t[session_file_names[i]] = eandt
    i = i + 1
```

```python
def CS_and_ITI_time(data, CS_s = 22, CS_e = 23, ITI_s = 15, ITI_e = 16):
    time = {}
    ITI_start = []
    ITI_end = []
    for row in data:
        if row[0] == ITI_s:
            ITI_start.append(row[1])
        if row[0] == ITI_e:
            ITI_end.append(row[1]) 

    CS_start = []
    CS_end = []
    for row in data:
        if row[0] == CS_s:
            CS_start.append(row[1])
        if row[0] == CS_e:
            CS_end.append(row[1])

    names = ['ITI_start', 'ITI_end', 'CS_start', 'CS_end']
    data = [ITI_start, ITI_end, CS_start, CS_end]
    i = 0
    for name in names:
        time[name] = data[i]
        i = i + 1
    
    return time
```

```python
i = 0
CS_ITI_times = {}
while i < len(session_file_names):
    CS_ITI = CS_and_ITI_time(e_and_t[session_file_names[i]], CS_s = 22, CS_e = 23, ITI_s = 15, ITI_e = 16)
    CS_ITI_times[session_file_names[i]] = CS_ITI
    i = i + 1
```

```python
def get_CS_npl(data, time, right_lp = 1, left_lp = 2, he = 6):
    
    CS_lp = 0
    CS_lp_prob = 0
    CS_lp_latency_ave = 0
    
    CS_lp_times = {}
    CS_lp_latency = {}
    
    lp_trial_counter = 0
    CS_lp_latency_sum = 0
    
    CS_he = 0
    CS_he_prob = 0
    CS_he_latency_ave = 0
    
    CS_he_times = {}
    CS_he_latency = {}
        
    he_trial_counter = 0
    CS_he_latency_sum = 0

    i = 0
    while i < len(time['CS_start']):
        j = 0
        for row in data:
            if row[1] >= time['CS_start'][i] and row[1] <= time['CS_end'][i]:
                if row[0] == right_lp:
                    CS_lp = CS_lp + 1
                    CS_lp_times[row[1]] = row[0]
                    if j < 1:
                        j = j + 1
                        lp_trial_counter = lp_trial_counter + 1
                        CS_lp_latency[row[1]] = row[1] - time['CS_start'][i]
                        CS_lp_latency_sum = CS_lp_latency_sum + CS_lp_latency[row[1]]
                if row[0] == left_lp:
                    CS_lp = CS_lp + 1
                    CS_lp_times[row[1]] = row[0]
                    if j < 1:
                        j = j + 1
                        lp_trial_counter = lp_trial_counter + 1
                        CS_lp_latency[row[1]] = row[1] - time['CS_start'][i]
                        CS_lp_latency_sum = CS_lp_latency_sum + CS_lp_latency[row[1]]
                if row[0] == he:
                    CS_he = CS_he + 1
                    CS_he_times[time['CS_start'][i]] = row[1]
                    if j < 1:
                        j = j + 1
                        he_trial_counter = he_trial_counter + 1
                        CS_he_latency[row[1]] = row[1] - time['CS_start'][i]
                        CS_he_latency_sum = CS_he_latency_sum + CS_he_latency[row[1]]
                
        i = i + 1
    
    CS_lp_prob = lp_trial_counter / len(time['CS_start'])
    if CS_lp_latency_sum == 0.0:
        CS_lp_latency_ave = 10
    else:
        CS_lp_latency_ave = CS_lp_latency_sum / lp_trial_counter
            
    CS_he_prob = he_trial_counter / len(time['CS_start'])
    if CS_he_latency_sum == 0.0:
        CS_he_latency_ave = 10
    else:
        CS_he_latency_ave = CS_he_latency_sum / he_trial_counter
    
    npl = {}
    names = ['CS_lp', 'CS_lp_times', 'CS_lp_prob', 'CS_lp_latency_ave', 'CS_lp_latency', 'CS_he', 'CS_he_times', 'CS_he_prob', 'CS_he_latency_ave', 'CS_he_latency']
    data = [CS_lp, CS_lp_times, CS_lp_prob, CS_lp_latency_ave, CS_lp_latency, CS_he, CS_he_times, CS_he_prob, CS_he_latency_ave, CS_he_latency]
    i = 0
    for name in names:
        npl[name] = data[i]
        i = i + 1
    
    return npl
```

```python
i = 0
npls = {}
while i < len(session_file_names):
    npl_temp = get_CS_npl(e_and_t[session_file_names[i]], CS_ITI_times[session_file_names[i]], right_lp = 1, left_lp = 2, he = 6)
    npls[session_file_names[i]] = npl_temp
    i = i + 1
```

```python
def get_PCA_score(data):
    if abs(data['CS_lp']) > 0 | abs(data['CS_he']) > 0:
        response_bias = (data['CS_lp'] - data['CS_he']) / (data['CS_lp'] + data['CS_he'])
    else:
        response_bias = 0
    prob_diff = data['CS_lp_prob'] - data['CS_he_prob']
    lat_score = (data['CS_he_latency_ave'] - data['CS_lp_latency_ave']) / 10
    PCA_score = (response_bias + prob_diff + lat_score) / 3
    
    scores = {}
    names = ['response_bias', 'prob_diff', 'lat_score', 'PCA_score']
    data = [response_bias, prob_diff, lat_score, PCA_score]
    i = 0
    for name in names:
        scores[name] = data[i]
        i = i + 1
        
    return scores
```
